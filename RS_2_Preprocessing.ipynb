```python
import pandas as pd

ratings = pd.read_csv(
    "data/raw/ratings.dat",
    sep="::",
    engine="python",
    names=["UserID", "MovieID", "Rating", "Timestamp"]
)

movies = pd.read_csv(
    "data/raw/movies.dat",
    sep="::",
    engine="python",
    encoding="latin-1",
    names=["MovieID", "Title", "Genres"]
)

users = pd.read_csv(
    "data/raw/users.dat",
    sep="::",
    engine="python",
    names=["UserID", "Gender", "Age", "Occupation", "ZipCode"]
)

print(ratings.shape, movies.shape, users.shape)
```

# Merge


```python

  data= ratings.merge(users,on="UserID")
  data= data.merge(movies,on="MovieID")
  data.head()
```

# Remove Sparse Movies

```python


movie_counts=ratings.groupby("MovieID").size()      # Count number of ratings per movie

valid_movies= movie_counts[movie_counts>4].index    # keep only movies with more than 4 ratings

ratings= ratings[ratings["MovieID"].isin(valid_movies)]   # filter dataset

print(ratings.shape)
```

# Rebuild dataset

```python

data = ratings.merge(users, on="UserID")
data = data.merge(movies, on="MovieID")
```

```python
data = data.drop_duplicates()
```

```python
import os

# Create Folder for filtered dataset
os.makedirs("data/processed", exist_ok=True)

# Save dataset in that folder
data.to_csv("data/processed/ratings_clean.csv", index=False)

# Save filtered ratings seperately
ratings.to_csv("data/processed/ratings_filtered.csv", index=False)

print("Saved processed datasets")
```
