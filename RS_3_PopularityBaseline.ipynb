# Leave-One-Out-Split

```python
ratings= pd.read_csv("data/processed/ratings_clean.csv")      # Load processed dataset

ratings = ratings.sort_values(["UserID", "Timestamp"])    # Sort by user and time (past-> future)

test = ratings.groupby("UserID").tail(1)     # Take last interaction as test
train = ratings.drop(test.index)             # Remove test from dataset

# Print dataset sizes
print("Train shape:",train.shape)
print("Test shape:",test.shape)
```

# Baseline Model- Popularity-Based Recommendation

```python
#Load Movies (Only for metadata-not used for training)
movies = pd.read_csv(
    "data/raw/movies.dat",
    sep="::",
    engine="python",
    encoding="latin-1",
    names=["MovieID", "Title", "Genres"]
)

# Count how many ratings each movie received in Train data (More ratings= more popular movie)
popular_movies = train.groupby("MovieID").agg(
    NumRatings=("Rating", "count")
)

# Sort movies by popularity
popular_movies = popular_movies.sort_values(
    by="NumRatings",
    ascending=False
)

# Merge with movies dataset to convert MovieID to title
popular_movies = popular_movies.merge(
    movies,
    on="MovieID"
)

# Get top 10 recommendations
top10 =popular_movies[
    ["Title","NumRatings"]
].head(10)

print(top10)

```
