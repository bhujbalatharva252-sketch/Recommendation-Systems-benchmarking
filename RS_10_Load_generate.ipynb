```python
# Import required libraries

import torch

from transformers import AutoTokenizer, AutoModelForCausalLM
```

# Load Model

```python
# Function to load the LLM model

def load_llm(model_name):

    # Load tokenizer
    tokenizer=AutoTokenizer.from_pretrained(model_name)

    # Load the actual language model
    model=AutoModelForCausalLM.from_pretrained(
        model_name,
        torch_dtype=torch.float16,   # Use less GPU memory
        device_map="auto"            # Automatically use available GPU
    )

    return tokenizer,model
```

# Generate Recommendation

```python
def generate_recommendations(prompt, tokenizer, model):

    # Convert text prompt into tokens that the model understands
    inputs=tokenizer(
        prompt,
        return_tensors="pt",
        truncation=True
    )

    # Move input to same device as model (GPU/CPU)
    inputs=inputs.to(model.device)


    # Generate response from the model
    outputs=model.generate(**inputs,
        max_new_tokens=300,   # Maximum length of generated response
        do_sample=False      # Deterministic output
    )

    # Convert tokens back into readable text
    response=tokenizer.decode(outputs[0],skip_special_tokens=True)

    return response
```

```python
# Example
# Choose model

model_name="mistralai/Mistral-7B-Instruct-v0.3"

# Load model
tokenizer, model = load_llm(model_name)

# Example prompt
prompt = """Recommend 5 science fiction movies
            - Do not provide explanations.
            - Do not provide descriptions."""

# Generate response
recommendations=generate_recommendations(prompt,tokenizer,model)

print(recommendations)
```
